###Import Libraries

###Use Request to Import API

```python
import requests
```

###Define URL and Params (parameters you want from the URL)

```python
url="https://api.coingecko.com/api/v3/coins/markets?"
import requests
param = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 250,
    "page": 1,
    }

response = requests.get(url=url, params=param)
if response.status_code==200:
    print('API successfully connected and data extracted')
else:
    print(f'error due to:{response.status_code}')
```

###Pass response as a JSON file so it is readable by Python

```python
data = response.json()
```

```python
print(data[0])
```

####Data Transformation before loading. The format in which the imported data type is structured, it is not consistent. Spark works only when the data schema type is consistent, hence, reason for transformation.

```python
import pandas as pd

df = spark.createDataFrame(
    pd.DataFrame(data)
)
```

```python
display(df)
```
